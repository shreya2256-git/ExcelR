```python
# Handling outliers in 'legs' using the IQR method
Q1 = zoo['legs'].quantile(0.25)
Q3 = zoo['legs'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
```

```python
# Cap the outliers
zoo['legs'] = zoo['legs'].clip(lower=lower_bound, upper=upper_bound)
```

```python
# Visualize the distribution of 'legs' after handling outliers
plt.figure(figsize=(10, 5))
sns.boxplot(x=zoo['legs'], palette='pastel')
plt.title('Boxplot of Legs (After Handling Outliers)')
plt.xlabel('Legs')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X = zoo.iloc[:,1:16]
Y = zoo.iloc[:,16]
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=1, stratify=Y)
```

```python
X_train
```

```python
num_folds=10
kfold= KFold(n_splits=10)
```

```python
model= KNeighborsClassifier(n_neighbors=3)
model.fit(X_train,Y_train)
```

```python
preds = model.predict(X_test)
pd.Series(preds).value_counts()
```

```python
pd.crosstab(Y_test,preds)
```

```python
np.mean(preds==Y_test)
```

```python
model.score(X_train,Y_train)
```

```python
print("Accuracy", accuracy_score(Y_test,preds)*100)
```

```python
results = cross_val_score(model, X, Y, cv=kfold)
```

```python
print(results.mean()*100)
```

```python
print(results.std()*100)
```

```python
n_neighbors = np.array(range(1,40))
param_grid = dict(n_neighbors=n_neighbors)
```

```python
model = KNeighborsClassifier()
grid = GridSearchCV(estimator=model, param_grid=param_grid)
grid.fit(X, Y)
```

```python
print(grid.best_score_)
print(grid.best_params_)
```

```python
k_values = np.arange(1,25)
train_accuracy = []
test_accuracy = []
```

```python
for i, k in enumerate(k_values):
    # k from 1 to 25(exclude)
    knn = KNeighborsClassifier(n_neighbors=k)
    # Fit with knn
    knn.fit(X_train,Y_train)
    #train accuracy
    train_accuracy.append(knn.score(X_train,Y_train))
    # test accuracy
    test_accuracy.append(knn.score(X_test,Y_test))
# Plot
plt.figure(figsize=[13,8])
plt.plot(k_values, test_accuracy, label = 'Testing Accuracy')
plt.plot(k_values, train_accuracy, label = 'Training Accuracy')
plt.legend()
plt.title('-value VS Accuracy')
plt.xlabel('Number of Neighbors')
plt.ylabel('Accuracy')
plt.xticks(k_values)
plt.savefig('graph.png')
plt.show()
print("Best accuracy is {} with K = {}".format(np.max(test_accuracy),1+test_accuracy.index(np.max(test_accuracy))))
```
